```python
import os
from PIL import Image
import numpy as np
from skimage import io
from random import randrange
from skimage.transform import rotate
import matplotlib.pyplot as plt
import cv2
from skimage.io import imread
```

```python
DHM_test_path = '/Users/behnazyazdani/Img_augmentation/Cell_img/'
U_all_tiff_files = []
N = 0
for file in os.listdir(DHM_test_path):
    if file.startswith(""):
        if file.endswith(".tif"):
            U_all_tiff_files.append(file)
            N+=1
        
U_all_names = [] 
for file in U_all_tiff_files: 
    U_all_names.append(file[:-4])
   
    
U_all_names = sorted(U_all_names, key = lambda x: (str(x[10:])))


U_all_sorted_tiff_files = [name+'.tif' for name in U_all_names]

U_all_sorted_directories = [ DHM_test_path + "/" + file for file in U_all_sorted_tiff_files ]
U_all_sorted_directories

Mask_test_path = '/Users/behnazyazdani/Img_augmentation/Cell_lbl/'
M_all_tiff_files = []
N = 0
for file in os.listdir(Mask_test_path):
    if file.startswith(""):
        if file.endswith(".tif"):
            M_all_tiff_files.append(file)
            N+=1
        #print(os.path.join(data_directory, file))
        
M_all_names = [] 
for file in M_all_tiff_files: 
    M_all_names.append(file[:-4])
  
    
M_all_names = sorted(M_all_names, key = lambda x: (str(x[10:])))
M_all_sorted_tiff_files = [name+'.tif' for name in M_all_names]
M_all_sorted_directories = [ Mask_test_path + "/" + file for file in M_all_sorted_tiff_files ]
print(M_all_sorted_directories)
U_all_sorted_directories
```

```python


#DHM_test_path = '/Users/behnazyazdani/Img_augmentation/Cell_img'
DHM_save_path = '/Users/behnazyazdani/Img_augmentation/test/img'
#Mask_test_path = '/Users/behnazyazdani/Img_augmentation/Cell_lbl'
Mask_save_path = '/Users/behnazyazdani/Img_augmentation/test/lbl'

angles = [0, 15, 30, 45, 60, 75, 90]
matrix = 128
sample = 10

# Mask_file_list = [f for f in os.listdir(Mask_test_path) if f[-3:]=="tif"]
# DHM_file_list = [f for f in os.listdir(DHM_test_path) if f[-3:]=="tif"]

for i in range(len(U_all_sorted_directories)):
    DHM_img = imread(str(U_all_sorted_directories[i]))
    img = np.array(DHM_img, dtype=np.uint8)
    mask = imread(str(M_all_sorted_directories[i])) 
    mask_img =np.array( mask , dtype=np.uint8)
    mask_img = mask_img>0.99
    for angle in angles:
        DHM_img_rt = rotate(img, angle=angle)
        mask_rt = rotate(mask_img,angle = angle)
        x, y = np.shape(DHM_img_rt)
        for j in range(sample):
            x1 = randrange(0, x - matrix)
            y1 = randrange(0, y - matrix)
            DHM_img_crop = DHM_img_rt[x1:x1 + matrix, y1:y1 + matrix]
            DHM_img_crop_8bit = (DHM_img_crop * 255.0 / DHM_img_crop.max()).astype(np.uint8)
            mask_crop = mask_rt[x1:x1 + matrix, y1:y1 + matrix]
            
            # Create filenames for the saved images
            dhm_filename = os.path.splitext(os.path.basename(U_all_sorted_directories[i]))[0] + "_" + str(j) + "_rt_" + str(angle) + ".tif"
            mask_filename = os.path.splitext(os.path.basename(M_all_sorted_directories[i]))[0] + "_Mask_" + str(j) + "_rt_" + str(angle) + ".tif"
            # Save images to the specified paths
            io.imsave(os.path.join(DHM_save_path, dhm_filename), DHM_img_crop_8bit)
            io.imsave(os.path.join(Mask_save_path, mask_filename), mask_crop)
   
```
